```python
import pandas as pd

df = pd.read_csv('vehicle.csv')

print(df.head(10))
```

```python
import matplotlib.pyplot as plt

vehicle_type_counts = df['VehicleType'].value_counts()

vehicle_type_counts.plot(kind = "bar", color='skyblue')
plt.title('Distribution of Vehicle Types')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.show()
```

```python
journey_purpose_counts = df['JourneyPurpose'].value_counts()

journey_purpose_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightskyblue', 'lightgreen','lightgrey'])
plt.title('Journey Purpose Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.hist(df['AgeVehicle'].dropna(),bins=20, color='orange', edgecolor='black')
plt.title('Distribution of Vehicle Age')
plt.xlabel('Vehicle Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['AgeVehicle'], df['PointImpact'], alpha=0.6, color='purple')
plt.title('Vehicle Age vs. Point of Impact')
plt.xlabel('Vehicle Age')
plt.ylabel('Point of Impact')
plt.show()
```

```python
lefthand_counts = df['LeftHand'].value_counts()

lefthand_counts.plot(kind='bar', color=['salmon', 'teal'])
plt.title('Left-Hand vs. Right-Hand Vehicles')
plt.xlabel('Hand Preference')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
propulsion_counts = df.groupby(['VehicleType', 'Propulsion']).size().unstack()

propulsion_counts.plot(kind='bar', stacked=True, color=['#ff9999','#66b3ff','#99ff99'])
plt.title('Vehicle Type by Propulsion')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.show()
```
